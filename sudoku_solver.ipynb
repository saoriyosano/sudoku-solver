# 1. Validate solutions

## 1.1. Function to validate

```python
def check_complete(grid):
    nums = list(range(1, 10))
    while len(nums) > 0:
        for num in grid:
            if num in nums:
                nums.remove(num)
            else:
                return False
    return True
```

```python
import itertools
```

```python
def sudoku_validator(grid):
    # check rows
    for row_num, row in enumerate(grid):
        if not check_complete(row):
            return False

    # check columns
    for index in range(9):
        column = [grid[i][index] for i in range(9)]
        if not check_complete(column):
            return False

    # check 3x3 grids
    num_x = 0
    num_y = 3
    for i in range(len(grid)):
        smallgrid = []
        index = 0
        for i in range(3):
            smallgrid.append(grid[index][num_x:num_y])
            index += 1
        smallgrid = list(itertools.chain(*smallgrid))
        if not check_complete(smallgrid):
            return False
        index += 3
        if index == 9:
            num_x += 3
            num_y += 3
        
    return True
```

# 2. Solve sudoku

```python
# find candidates for the specified position

def get_candidates(state, rownum, colnum):
        
    candidates = list(range(1, 10))

    cur_row = [item for item in state[rownum] if item > 0] # row
    cur_column = [state[i][colnum] for i in range(9) if state[i][colnum] > 0] # column
    # mini grid
    cur_rownums = [rownum // 3 * 3, rownum // 3 * 3 + 1, rownum // 3 * 3 + 2]
    cur_colnums = [colnum // 3 * 3, colnum // 3 * 3 + 1, colnum // 3 * 3 + 2]
    cur_minigrid = [state[cur_rownum][cur_colnum] for cur_colnum in cur_colnums for cur_rownum in cur_rownums if state[cur_rownum][cur_colnum] > 0]
    
    existing = set(cur_row + cur_column + cur_minigrid)
    candidates = [candidate for candidate in candidates if candidate not in existing]
    print(f"found candidates {candidates} for [{rownum}, {colnum}]")
    
    return candidates
                        
```

```python
def search(state):
    
    if sudoku_validator(state):
        return True
    
    for rownum, row in enumerate(state):
            for colnum, item in enumerate(row):
                if item == 0:
                    for candidate in get_candidates(state, rownum, colnum):
                        state[rownum][colnum] = candidate
                        is_solved = search(state)
                        if is_solved:
                            return True
                        else:
                            state[rownum][colnum] = 0
                    return False
    return True
    
    # position, candidates = get_candidates(state)
    
    
    # for candidate in candidates:
    #     state[position[0]][position[1]] = candidate
    #     print(f"added {candidate} to position {position}")
    #     search(state)
    #     state[position[0]][position[1]] = 0
    #     print(f"removed number from {position}")
    #     counter += 1
    #     if counter == 10:
    #         break
```
